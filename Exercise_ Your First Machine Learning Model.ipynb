```python
#importing the pandas
import pandas as pd
#importing the scikit learn lib
from sklearn.tree import DecisionTreeRegressor

#declaring the file path to the dataframe
melbourne_file_path = '../input/melbourne-housing-snapshot/melb_data.csv'

#reading the data using pandas
melbourne_data = pd.read_csv(melbourne_file_path) 

#printing the column names of the data frame
melbourne_data.columns

#dropping the not available A.K.A. null values columns
melbourne_data = melbourne_data.dropna(axis=0)

#setting the y axix variable, this is the variable that we need to make predictions of
y = melbourne_data.Price

#setting the x axix variables, this is the variables that we use to make the predictions
prediction_features = ['Rooms','Bathroom','Landsize','Lattitude','Longtitude']
x = melbourne_data[prediction_features]  #this should be inside a array/list

#display the filtered dataset/dataframe
x.describe()
x.head()

#creating the model,
model = DecisionTreeRegressor(random_state = 1)

#fit the model to the data
model.fit(x,y)

#predicting the data
print("making predictions for the following 5 houses")
print(x.head())
print("predictiosn are : ")
print(model.predict(x.head()))



```
